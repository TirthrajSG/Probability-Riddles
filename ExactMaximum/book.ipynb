# Problem Statement: 

### 8 fare die are rolled, find the probability that maximum of the rolls is exactly 5. 


## Import Libraries 

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Create a function to calculate probability according to the problem

```python
def find_prob(n = 100):
    fav_outcomes = 0
    for _ in range(n):
        max_roll = max(np.random.randint(1,7, size=8))
        if max_roll == 5:
            fav_outcomes += 1
    return fav_outcomes/n
```

## Create a function to compute running mean, to show convergence using a graph

```python
def running_mean(arr):
    new_arr = []
    total = 0
    for i, x in enumerate(arr, 1):
        total += x
        new_arr.append(total / i)
    return new_arr
```

## Compute the probability

```python
probs = [find_prob() for _ in range(1000)]
running_probs = running_mean(probs)
actual_prob = (5**8 - 4**8)/6**8
```

## Plot the probability

```python
plt.plot(running_probs, label="Simulated Probability")
plt.axhline(y=actual_prob, color='red', linestyle='--', label='Expected value')
plt.legend()
plt.show()
```
