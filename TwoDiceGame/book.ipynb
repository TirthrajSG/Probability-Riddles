# Problem Statement:
Two players take turns rolling two six-sided dice. Player A goes first, followed by player B. If player A rolls a sum of 6, they win. If player B rolls a sum of 7, they win. If neither rolls their desired value, the game continues until someone wins. What is the probability that player A wins?

### Import Libraries

```python
import random
import matplotlib.pyplot as plt
```

### Define Running Mean Function

```python
def running_mean(arr):
    new_arr = []
    total = 0
    for i, x in enumerate(arr, 1):
        total += x
        new_arr.append(total / i)
    return new_arr
```

### Define Caculate Probability Function

```python
def find_prob(n=10000):
    no_of_wins_A = 0
    for i in range(n):
        while 1:
            a1 = random.randint(1,6)
            a2 = random.randint(1,6)

            if a1+a2 == 6:
                no_of_wins_A += 1
                break
            
            b1 = random.randint(1,6)
            b2 = random.randint(1,6)

            if b1+b2 == 7:
                break
    return no_of_wins_A/n
```

### Calculate Probability

```python
no_of_simulations = 1000
probs = [find_prob() for _ in range(no_of_simulations)]
running_probs = running_mean(probs)
actual_prob = 30/61
```

### Plot the Simulated Probability

```python
plt.plot(running_probs, label="Simulated Probability")
plt.axhline(y=actual_prob, color='red', linestyle='--', label='Expected value')
plt.legend()
plt.show()
```
