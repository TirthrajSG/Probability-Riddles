## We will calculate the Expection of arc length containing (0,1), when 3 points are chosen on a unit circle centered at origin.

### **Step 1:** Import libraries.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### **Step 2:** Define a function which return arc length.

```python
def find_arc_len():
    theta = np.random.uniform(0, 2*np.pi, 3)
    return 2*np.pi - max(theta) + min(theta)
```

### **Step 3:** Define a function to calculate running mean.

```python
def running_mean(arr):
    new_arr = []
    total = 0
    for i, x in enumerate(arr, 1):
        total += x
        new_arr.append(total / i)
    return new_arr
```

### **Step 4:** Find expectation.

```python
expectations = []
for _ in range(1000):
    arc_lens = [find_arc_len() for __ in range(100)]
    expectation = np.mean(arc_lens)
    expectations.append(expectation)

print(np.mean(expectations))
```

### **Step 5:** Show convergence by plot.

```python
running_mean_exp = running_mean(expectations)
actual_expectation = np.pi

plt.plot(running_mean_exp, label="Simulated Probability")
plt.axhline(y=actual_expectation, color='red', linestyle='--', label='Expected value')
plt.legend()
plt.show()
```
